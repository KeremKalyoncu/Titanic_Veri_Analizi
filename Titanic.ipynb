# Hayatta KalÄ±p Kalmama Tahmini

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df
```

```python
df.drop(columns=["Name","Ticket","Embarked","Fare"],inplace=True)
```

```python
df
```

```python
df["Sex"] = pd.get_dummies(df["Sex"],drop_first=True)
```

```python
df
```

```python
x = df.drop("Survived",axis=1)
y = df["Survived"]

x_train , x_test, y_train , y_test = train_test_split(x,y,train_size=0.75,random_state=53)
```

```python
log = LogisticRegression()
model = log.fit(x_train,y_train)
model.score(x_train,y_train)
```

```python
# 1 olmasÄ± iyi gibi ama overfitting olma durumu olabilir bi bakalÄ±m
```

```python
model.predict([[1,0,39,0,0]])
```
